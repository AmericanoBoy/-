Задача 2.6
Посмотрите распределение возраста ( age ): гистограмма ( plt.hist ) и
базовые описательные характеристики ( min , max , mean , median , std )
Разбейте пассажиров на возрастные группы (бины): 0-12, 13-18, 19-30,
31-50, 51+. Для каждой группы посчитайте долю выживших.
Постройте линию ( plt.plot ) - возрастная группа по оси X и процент
выживших - по Y.
Подсказка: изучите в документации matplotlib
, что можно передавать в
аргумент bins
функции hist
.

```python
bin, lab = [0,12,18,30,50,80], ['0=<12','13=<18', '19=<30', '31=<50', '>=51']
titan_pd['AgeGroup'] = pd.cut(titan_pd['Age'], bins= bin, labels=lab, right=True)
count_every_age_group = titan_pd.groupby(['AgeGroup']).size()
count_survived = titan_pd[titan_pd['Survived']==1].groupby(['AgeGroup']).size()
percent_survived = ((count_survived/count_every_age_group)*100).round(2)
percent_survived
```

```python

plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.title('КОЛ_ВО ПАССАЖИРОВ КАЖДОЙ ВОЗРАСТНОЙ ГРУППЫ')
plt.bar(lab,count_every_age_group)


plt.subplot(1,2,2)
plt.plot([0, 10, 30, 50, 80], percent_survived)
plt.title('ПРОЦЕНТ ВЫЖИВШИХ СРЕДИ КАЖДОЙ ВОЗРАСТНОЙ ГРУППЫ')
```

Задача 2.7
Добавьте колонку FamilySize = SibSp + Parch + 1 (1 - сам пассажир).
Посчитайте, как меняется выживаемость в зависимости от FamilySize .
Постройте график ( plt.bar/plt.hist ) с выживаемостью для каждого
размера семьи.
Сформируйте булевую колонку IsAlone = FamilySize == 1 и сравните
выживаемость одиночек и не-одиночек.

```python
titan_pd['FamilySize'] = titan_pd['SibSp'] + titan_pd['Parch'] + 1
res = titan_pd.groupby(['Survived','FamilySize']).size()
titan_pd['IsAlone'] = titan_pd['FamilySize'] == 1
survived_ones_and_not_ones = titan_pd.groupby(['IsAlone', 'Survived']).size().unstack(fill_value=0)
res,survived_ones_and_not_ones
```

```python
plt.hist(titan_pd[titan_pd['Survived']==1]['FamilySize'],alpha=1,label='ВЫЖИВШИЕ')
plt.hist(titan_pd[titan_pd['Survived']==0]['FamilySize'], alpha=0.5,label='ПОГИБШИЕ')
plt.title('ЗАВИСИМОСТЬ ВЫЖИВАНИЯ ОТ КОЛ_ВА ЧЛЕНОВ СЕМЬИ')
plt.legend()
plt.xlabel('КОЛ_ВО ЧЛЕНОВ СЕМЬИ')
plt.ylabel('КОЛ_ВО ПАССАЖИРОВ')
plt.show()
#ВЫВОД: большее кол_во выживших где кол_во членов семьи незначительно
```

Задача 2.9 (ПО ЖЕЛАНИЮ!)
Постройте несколько вызовов plt.scatter подряд: точки выживших
( Survived==1 ) одним цветом, невыживших ( Survived==0 ) другим
(внутри каждой группы используйте разные маркеры или
прозрачность для полов (Sex));
Размер точки масштабируйте пропорционально Fare (например, s =
np.sqrt(Fare)*k ), чтобы показать влияние стоимости билета.
Подпишите оси, добавьте легенду и подписи для ключевых
наблюдений.
Сформулируйте краткий аналитический вывод: как связаны Age , Fare ,
Sex и вероятность выживания?




```python
color_survived = [['green','red'][sex=='female']for sex in titan_pd[titan_pd['Survived']==False]['Sex']]
color_dead =[['green','red'][sex=='female']for sex in titan_pd[titan_pd['Survived']==True]['Sex']]
plt.scatter(titan_pd[titan_pd['Survived'] == True]['Age'], titan_pd[titan_pd['Survived'] == 1]['Fare'], color='white',
                edgecolors=color_dead, label='ВЫЖИВШИЕ')
plt.scatter(titan_pd[titan_pd['Survived'] == False]['Age'], titan_pd[titan_pd['Survived'] == 0]['Fare'], color='black',
             edgecolors=color_survived, label = 'ПОГИБШИЕ')
plt.legend()
plt.xlabel('ВОЗРАСТ')
plt.ylabel('СТОИМОСТЬ БИЛЕТА')
plt.title('ЗАВИСИМОСТЬ ВЫЖИВАНИЯ ОТ ПОЛА И ЦЕНЫ БИЛЕТА')
plt.show()
#ВЫВОД: чем ниже стоимость билета процент выживших мужчин и женщин одинаков по возрасту, с возрастанием цены билета процент
#выживших женщин заметно выше в ценовой категории от 2 класса и выше
```
